## Plotting

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#ex1: Plotting x and y points
xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
#ex2:Plotting Without Line
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

```python
#ex3: Multiple Points
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

# Marker Size

```python
#ex9
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20)
plt.show()
```

# Marker Colour
'r'    Red    
'g'    Green    
'b'    Blue    
'c'    Cyan    
'm'    Magenta    
'y'    Yellow    
'k'    Black    
'w'    White

```python
#ex10: marker edge colour
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')
plt.show()
```

```python
#ex11: marker face colour
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r')
plt.show()
```

```python
#ex12: both edge and face
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r', mfc = 'r')
plt.show()
```

# Lines
'solid' (default)    '-'    
'dotted'    ':'    
'dashed'    '--'    
'dashdot'    '-.'

```python
#ex13:
ypoints = np.array([3, 8, 1, 10])

# if you use marker argument name, you can only use markers not lines together
plt.plot(ypoints, 'o:')
plt.show()
```

```python
#ex14: line width
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linewidth = '20.5')
plt.show()
```

```python
#ex14: multi lines
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)

plt.show()
```

# Draw two lines
